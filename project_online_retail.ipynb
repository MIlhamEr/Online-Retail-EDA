# Data Cleaning

## Dropping Missing Values

```python
df = df.dropna()
df.isnull().sum()
```

## Dropping cancelled products from invoice

```python
df = df[~df.InvoiceNo.str.contains('C',na=False)]
print("Dataset bersih dari cancelled products")
```

**!!! NOTE !!!**

~ adalah operator negasi. Ini membalikkan hasil dari kondisi di atas. Artinya, kita akan memilih baris-baris yang tidak memenuhi kondisi (yaitu, tidak mengandung karakter 'C' dalam kolom 'InvoiceNo').


## Dropping Duplicated Data

```python
df = df.drop_duplicates(keep="first")
```

**!!! NOTE !!!**

"first" berarti kita akan tetap mempertahankan baris pertama yang ditemui jika ada beberapa baris yang memiliki nilai yang sama di semua kolom. Dengan kata lain, hanya baris pertama dari setiap rangkaian duplikat yang akan tetap ada dalam DataFrame.

## Cleaning Outliers

outlier: nilai yang berbeda secara signifikan dari mayoritas data.

**outlier_thresholds**

fungsi yang digunakan untuk menghitung batas atas (upper limit) dan batas bawah (lower limit) untuk outlier berdasarkan sebuah variabel tertentu dalam DataFrame.

**replace_with_threshold**

fungsi yang digunakan untuk menggantikan nilai outlier dalam DataFrame dengan batas atas dan batas bawah yang telah dihitung sebelumnya.

Mendefinisikan fungsi untuk menghapus data pencilan (outliers) berdasarkan batasannya

```python
def outlier_thresholds(dataframe, variable):
    quartile1 = dataframe[variable].quantile(0.01)
    quartile3 = dataframe[variable].quantile(0.99)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return up_limit, low_limit

def replace_with_threshold(dataframe, variable):
    up_limit, low_limit = outlier_thresholds(dataframe, variable)
    dataframe.loc[(dataframe[variable] < low_limit), variable] = low_limit
    dataframe.loc[(dataframe[variable] > up_limit), variable] = up_limit
```

Observasi sebelum menghapus outlier

```python
f, ax = plt.subplots(2, figsize = (15,5))
col_list = ["Quantity","UnitPrice"]
for i in range(0,2):
    ax[i].boxplot(df[col_list[i]], flierprops = dict(marker = "s", markerfacecolor = "red"), vert = 0)
plt.show()
```

Menerapkan fungsi untuk menghapus outlier

```python
replace_with_threshold(df, "Quantity")
replace_with_threshold(df, "UnitPrice")
```

Observasi setelah menghapus outlier

```python
f, ax = plt.subplots(2,1, figsize = (15,5))
for i in range(0,2):
    ax[i].boxplot(df[col_list[i]],flierprops = dict(marker = "s", markerfacecolor = "red"), vert = 0)
plt.show()
```

## Sales Pattern

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['InvoiceDate']
```

### Daily Sales

```python
daily_sales = df.groupby(df['InvoiceDate'].dt.date)['Quantity'].sum()
daily_sales.head()
```

```python
plt.figure(figsize=(12, 6))
daily_sales.plot(color='darkblue')
plt.title("Daily Sales")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.grid()
plt.show()
```

```python
daily_sales_diff = daily_sales.diff().dropna()
daily_sales_diff_sorted = daily_sales_diff.sort_values(ascending=False)
daily_sales_diff_sorted_head = daily_sales_diff_sorted.head(10)
daily_sales_diff_sorted_tail = daily_sales_diff_sorted.tail(10)
```

```python
plt.figure(figsize=(72, 6))
daily_sales_diff_sorted.plot(kind='bar', color='darkblue')
plt.title("Daily Sales Increase (From the Largest)")
plt.xlabel("Date")
plt.ylabel("Sales Increases")
plt.grid(axis = 'x')
plt.show()
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
daily_sales_diff_sorted_head.plot(kind='barh', color='darkblue')
plt.title("Daily Sales Increase (Largest)")
plt.xlabel("Sales Increases")
plt.ylabel("Date")
plt.grid(axis = 'x', linestyle="--", color="lightgray")

plt.subplot(1, 2, 2)
daily_sales_diff_sorted_tail.plot(kind='barh', color='darkblue')
plt.title("Daily Sales Decline (Least)")
plt.xlabel("Sales Decline")
plt.ylabel("Date")
plt.grid(axis = 'x', linestyle="--", color="lightgray")

plt.tight_layout()
plt.show()
```
